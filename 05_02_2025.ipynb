```python
import re
def clean_text(text):
  text=re.sub(r'[^a-zA-Z\']','',text)
  text=re.sub(r'[^\x00-\x7F]+','',text)
  text=text.lower()
  return text
```

```python
dataset['clean_dataset']=dataset.tweet.apply(lambda x:clean_text(x))
```

```python
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords

len(stopwords.words('english'))
```

```python
stop = stopwords.words('english')
```

```python
def gen_freq(text):
    word_list = []

    for tw_words in text.split():

        word_list.extend(tw_words)


    word_freq = pd.Series(word_list).value_counts()


    word_freq = word_freq.drop(stop, errors='ignore')

    return word_freq
```

```python
def any_neg(words):
  for words in words:
    if words in ['n','no','non','not'] or re.search(r"\wn't",words):
      return 1
    else:
      return 0
```

```python
def any_rare(words, rare_100):

    for word in words:

        if word in rare_100:

            return 1

        else:

            return 0
```

```python
def is_question(words):

    for word in words:

        if word in ['when', 'what', 'how', 'why', 'who', 'where']:

            return 1

        else:

            return 0
```

```python
word_freq = gen_freq(dataset.clean_dataset.str) # Change 'clean_text' to 'clean_dataset'



rare_100 = word_freq[-100:]

dataset['word_count'] = dataset.clean_dataset.str.split().apply(lambda x: len(x)) # Change 'clean_text' to 'clean_dataset'


dataset['any_neg'] = dataset.clean_dataset.str.split().apply(lambda x: any_neg(x)) # Change 'clean_text' to 'clean_dataset'


dataset['is_question'] = dataset.clean_dataset.str.split().apply(lambda x: is_question(x)) # Change 'clean_text' to 'clean_dataset'


dataset['any_rare'] = dataset.clean_dataset.str.split().apply(lambda x: any_rare(x, rare_100)) # Change 'clean_text' to 'clean_dataset'


dataset['char_count'] = dataset.clean_dataset.apply(lambda x: len(x)) # Change 'clean_text' to 'clean_dataset'
```
